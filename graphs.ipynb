```python
import pylab
import pandas as pd
import numpy as np
import seaborn
```

```python
df = pd.read_csv('sfbarf_hackathon/regular.csv')
df_more_lux = pd.read_csv('sfbarf_hackathon/more_luxury.csv')
df_more_chp = pd.read_csv('sfbarf_hackathon/more_cheap.csv')
df.columns
```

```python
pallete = seaborn.color_palette("deep", 7)

pylab.figure(figsize=(10,10))
no_house = df[np.isfinite(df.apt_score) == False]
has_house = df[np.isfinite(df.apt_score) == True]

pylab.scatter(no_house.min_score, no_house.max_rent, alpha=0.6, color=pallete[2], label='Got no home')
pylab.scatter(has_house.min_score, has_house.max_rent, alpha=0.6, color=pallete[0], label='Got a home')
pylab.xlabel('Minimum Acceptable Housing Score', fontsize=20)
pylab.ylabel('Maximum Home Price', fontsize=20)
pylab.ylim((-10,4e4))
pylab.xlim((0.5, 5.5))
pylab.yticks()
pylab.show()
```

```python
pylab.figure(figsize=(10,10))
rents = has_house.rent
max_rents = df.max_rent

pylab.hist(max_rents, bins=200, color=pallete[1], alpha=0.6, label='Got no home')
pylab.hist(rents, bins=30, color=pallete[0], alpha=0.9, label='Got no home')
pylab.xlim((0, 3e4))
pylab.show()
```

```python
pylab.figure(figsize=(10,10))
no_house = df[np.isfinite(df.apt_score) == False]
pylab.hist(no_house.max_rent, bins=100, color=pallete[i+1], alpha=0.8, label='Got no house')
for i in np.arange(5):
    q_slice = df[df.apt_score == i+1]
    label = 'Got house with score %d' % (i + 1)
    pylab.hist(q_slice.max_rent, bins=50, color=pallete[i], label=label)
pylab.legend(loc=2, fontsize=14)
pylab.xscale('log')
pylab.xlabel('Maximum Acceptable Rental Price', fontsize=20)
pylab.ylabel('Number of People', fontsize=20)
pylab.ylim((0,800))
pylab.show()
```

```python
pylab.figure(figsize=(10,10))
no_house = df_more_lux[np.isfinite(df_more_lux.apt_score) == False]
pylab.hist(no_house.max_rent, bins=100, color=pallete[i+1], alpha=0.8, label='Got no house')
for i in np.arange(5):
    q_slice = df_more_lux[df.apt_score == i+1]
    label = 'Got house with score %d' % (i + 1)
    pylab.hist(q_slice.max_rent, bins=50, color=pallete[i], label=label)
pylab.legend(loc=2, fontsize=14)
pylab.xscale('log')
pylab.xlabel('Maximum Acceptable Rental Price', fontsize=20)
pylab.ylabel('Number of People', fontsize=20)
pylab.ylim((0,800))

pylab.show()
```

```python
pylab.figure(figsize=(10,10))
no_house_reg = df[np.isfinite(df.apt_score) == False]
no_house_more_lux = df_more_lux[np.isfinite(df_more_lux.apt_score) == False]

pylab.hist(no_house_reg.max_rent, bins=100, color=pallete[0], alpha=0.6, label='Regular Scenario')
pylab.hist(no_house_more_lux.max_rent, bins=100, color=pallete[3], alpha=0.7, label='More luxury apartments')
pylab.xscale('log')
pylab.xlabel('Maximum Acceptable Rental Price', fontsize=20)
pylab.ylabel('Number of People', fontsize=20)
pylab.legend(loc=2)
pylab.show()
```
